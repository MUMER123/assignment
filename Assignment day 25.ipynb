```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
```

```python
d = load_digits()
# input variables(x)
d.data.shape
x= d.data
```

```python
# output variables(y)
d.target.shape
y=d.target
```

```python
# plotting some pictures from the dataset

plt.figure(figsize=(20,4))
for index,(image,label) in enumerate(zip(d.data[0:10],d.target[0:10])):
    plt.subplot(1,10,index + 1)
    plt.imshow(np.reshape(image,(8,8)),cmap=plt.cm.gray)
    plt.title("Training: %i\n" % label,fontsize=20)
```

```python
# split the data
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print("Train input data : ",X_train.shape)
print("Train output data : ",y_train.shape)
print("Test input data : ",X_test.shape)
print("Test output data : ",y_test.shape)
```

```python
# model train
from sklearn.linear_model import LogisticRegression
model = LogisticRegression().fit(X_train,y_train)
```

```python
# prediction
model.predict(X_test[0:10])
```

```python
pred = model.predict(X_test)
pred
```

```python
# Accuracy test
score = model.score(X_test,y_test)
print("The accuracy score is : ",score)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,pred)
cm
```

```python
# plotting the confusion matrix
import seaborn as sns 
plt.figure(figsize=(9,9))
sns.heatmap(cm,annot = True,fmt='.3f',linewidth = .5,square = True,cmap = "Spectral")
plt.ylabel("Actual Output",color = "white")
plt.xlabel("Predicted output",color ='white')
all_sample_title = "Accuracy score:{0}".format(score)
plt.title(all_sample_title,size = 15,color = 'white')
```

```python
index = 0
misclassifiedindeces = []
for label,predict in zip(y_test, pred):
    if label!=predict:
        misclassifiedindeces.append(index)
        index+=1


print(misclassifiedindeces)
```

```python
# plotting misclassified indexes
plt.figure(figsize=(60,20))
for plotindex,badindex in enumerate(misclassifiedindeces[0:5]):
    plt.subplot(1,5,plotindex+1)
    plt.imshow(np.reshape(X_test[badindex],(8,8)),cmap=plt.cm.gray)
    plt.title("Predicted: {},\n Actual: {}".format(pred[index],y_test[badindex]), fontsize =50,color = 'white')
```
