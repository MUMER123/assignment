## DATA NORMALIZATION

```python
ks=sns.load_dataset("titanic")
ks.head()
```

```python
k=ks[['age','fare']]
k.head()
```

1. ## SIMPLE FEATURE SCALING(x=x/x.max())

```python
k['fare']=k['fare']/k['fare'].max()
k.head()
```

```python
k['age']=k['age']/k['age'].max()
k.head()
```

## 2. MIN MAX METHOD(X=(X-X.MIN)/(X.MAX-X.MIN))

```python
ks['fare']=(ks['fare']-ks['fare'].min())/(ks['fare'].max()-ks['fare'].min())
ks['age']=(ks['age']-ks['age'].min())/(ks['age'].max()-ks['age'].min())
ks[['age','fare']].head()
```

## 3. Z SCORE METHOD(X=(X-X.MEAN()/(X.STD())))

```python
t=sns.load_dataset('titanic')
t[['age','fare']].head()
```

```python
t['fare']=(t['fare']-t['fare'].mean())/t['fare'].std()
t['age']=(t['age']-t['age'].mean())/t['age'].std()
t.head()
```

## 4. LOG TRANSFORMATION(X=NP.LOG(X))

```python
d=sns.load_dataset("titanic")
d[['age','fare']].head()
```

```python
d['fare']=np.log(d['fare'])
d['age']=np.log(d['age'])
d.head()
```
