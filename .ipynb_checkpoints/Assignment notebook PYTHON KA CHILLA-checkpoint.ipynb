# Introduction To Graphs

### Bar plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=True)
titanic = sns.load_dataset("titanic")
sns.catplot(x="sex",y="survived",hue="class",kind="bar",data=titanic)
plt.show()
```

### Count Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=True)
titanic = sns.load_dataset("titanic")
p1=sns.countplot(x="sex",hue="class", data=titanic)
p1.set_title("plot for counting")
plt.show()
```

### Scatter plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=True)
titanic = sns.load_dataset("titanic")
g=sns.FacetGrid(titanic, row="sex", hue="alone")
g=(g.map(plt.scatter, "age" , "fare").add_legend())
plt.show()
```

### Plotting a Graph by Importing your own data file

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
chilla = pd.read_csv("data.csv")
print(chilla)

sns.set_theme(style="ticks",color_codes=True)
p = sns.countplot(x="Gender", hue="AREA", data=chilla)
plt.show()
 
```

#### To add title

```python
import seaborn as sns
import matplotlib.pyplot as plt
f=sns.load_dataset("iris")
f
sns.lineplot(x="sepal_length",y="sepal_width",data=f)
plt.title("GRAPH BETWEEN SEPAL LENGTH AND SEPAL WIDTH")
plt.show()
```

#### To add limits

```python
import seaborn as sns
import matplotlib.pyplot as plt
f=sns.load_dataset("iris")
f
sns.lineplot(x="sepal_length",y="sepal_width",data=f)
plt.title("GRAPH BETWEEN SEPAL LENGTH AND SEPAL WIDTH")
plt.xlim(3)
plt.ylim(2)
plt.show()
```

#### To add styles

#### To change size of plot

#### Using estimator

```python
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import mean 
f=sns.load_dataset("titanic")
f
sns.barplot(x="sex",y="fare",hue="class",data=f,estimator=mean)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import mean 
f=sns.load_dataset("titanic")
f
sns.barplot(x="sex",y="fare",hue="class",data=f,estimator=mean,saturation=0.2)
plt.show()
```

#### Showing mean in boxplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style(style="whitegrid")
f=sns.load_dataset("tips")
sns.boxplot(x="day",y="tip",data=f ,showmeans=True,meanprops={"marker":".","markersize":"10","markeredgecolor":"blue"})
plt.show()
```

#### Changing labels and title

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style(style="whitegrid")
f=sns.load_dataset("tips")
sns.boxplot(x="day",y="tip",data=f ,showmeans=True,meanprops={"marker":".","markersize":"10","markeredgecolor":"blue"})
plt.xlabel("Week Days",size=15)
plt.ylabel("Tip given",size=15)
plt.title("Resturaunt list",size=20,weight="bold")
plt.show()
```
