```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np

```

```python
df=sns.load_dataset("iris")
df.head()
```

```python
x=df.iloc[ : ,:-1]
y=df.iloc[ : ,-1:]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

model = DecisionTreeClassifier().fit(x,y)
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(figsize = (4,4), dpi=300)
plot_tree(model,filled = True,feature_names=fn,class_names=cn);
plt.title("Decision tree of iris trained model",color="white")
plt.show()
```

## To save the plot in png

```python
fig.savefig("tree.png")
```

## To save as tif


```python
fig.savefig("tree.tif")
```

## To save as pdf

```python
fig.savefig("tree.pdf")
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
model = DecisionTreeClassifier().fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```
